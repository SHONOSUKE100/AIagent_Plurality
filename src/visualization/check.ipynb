```python
import sqlite3
```

```python
con = sqlite3.connect("../../results/20251214-061105/simulation.db")
```

```python
cur = con.cursor()
```

```python
# 1. テーブル一覧を取得
cur.execute("SELECT name FROM sqlite_master WHERE type='table'")
tables = cur.fetchall()

# 2. 結果を表示
print("データベース内のテーブル一覧:")
for table_name, in tables:
    print(f"- {table_name}")
```

```python
if not tables:
    print("\nテーブルが見つかりませんでした。")
else:
    for table_name, in tables:
        print(f"\n=== テーブル: {table_name} ===")

        schema_rows = cur.execute(f"PRAGMA table_info('{table_name}')").fetchall()
        column_names = [column_info[1] for column_info in schema_rows]
        print("カラム一覧:")
        for cid, name, col_type, notnull, default, pk in schema_rows:
            print(f"  - {name} ({col_type}) NOT NULL={bool(notnull)} PK={'Yes' if pk else 'No'}")

        row_count = cur.execute(f"SELECT COUNT(*) FROM '{table_name}'").fetchone()[0]
        print(f"レコード件数: {row_count}")

        sample_rows = cur.execute(f"SELECT * FROM '{table_name}' LIMIT 5").fetchall()
        if sample_rows:
            print("サンプル (先頭5件):")
            print("  | ".join(column_names))
            for sample in sample_rows:
                formatted = [str(value) for value in sample]
                print("  | ".join(formatted))
        else:
            print("サンプル (先頭5件): データなし")
```
